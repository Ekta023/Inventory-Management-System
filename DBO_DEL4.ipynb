```python
import mysql.connector
from datetime import datetime

db = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Ludo@2024",
    database="inventory_management"
)
cursor = db.cursor()

def create_record(entity):
    try:
        if entity == "Product":
            cursor.execute("SELECT MAX(PID) FROM Product")
            max_pid = cursor.fetchone()[0]
            pid = max_pid + 1 if max_pid else 1
            pname = input("Enter Product Name: ")
            pcategory = input("Enter Product Category: ")
            punit_price = float(input("Enter Unit Price: "))
            pstock_quantity = int(input("Enter Stock Quantity: "))
            query = "INSERT INTO Product (PID, Pname, Pcategory, Punit_price, Pstock_quantity) VALUES (%s, %s, %s, %s, %s)"
            values = (pid, pname, pcategory, punit_price, pstock_quantity)
        elif entity == "Supplier":
            cursor.execute("SELECT MAX(SUPID) FROM Supplier")
            max_supid = cursor.fetchone()[0]
            supid = max_supid + 1 if max_supid else 1
            sname = input("Enter Supplier Name: ")
            saddress = input("Enter Supplier Address: ")
            scontact_info = input("Enter Supplier Contact Info: ")
            query = "INSERT INTO Supplier (SUPID, Sname, Saddress, Scontact_info) VALUES (%s, %s, %s, %s)"
            values = (supid, sname, saddress, scontact_info)
        elif entity == "Order":
            cursor.execute("SELECT MAX(OID) FROM `Order`")
            max_oid = cursor.fetchone()[0]
            oid = max_oid + 1 if max_oid else 1
            order_date = input("Enter Order Date (YYYY-MM-DD): ")
            supid = int(input("Enter Supplier ID: "))
            ototal_amount = float(input("Enter Total Amount: "))
            query = "INSERT INTO `Order` (OID, order_date, SUPID, Ototal_amount) VALUES (%s, %s, %s, %s)"
            values = (oid, order_date, supid, ototal_amount)
        elif entity == "Order_Details":
            cursor.execute("SELECT MAX(ODID) FROM Order_Details")
            max_odid = cursor.fetchone()[0]
            odid = max_odid + 1 if max_odid else 1
            oid = int(input("Enter Order ID: "))
            pid = int(input("Enter Product ID: "))
            odquantity = int(input("Enter Quantity: "))
            odunit_price = float(input("Enter Unit Price: "))
            query = "INSERT INTO Order_Details (ODID, OID, PID, ODquantity, ODunit_price) VALUES (%s, %s, %s, %s, %s)"
            values = (odid, oid, pid, odquantity, odunit_price)
        elif entity == "Customer":
            cursor.execute("SELECT MAX(CID) FROM Customer")
            max_cid = cursor.fetchone()[0]
            cid = max_cid + 1 if max_cid else 1
            cname = input("Enter Customer Name: ")
            caddress = input("Enter Customer Address: ")
            email = input("Enter Email: ")
            phone = input("Enter Phone: ")
            query = "INSERT INTO Customer (CID, Cname, Caddress, Email, Phone) VALUES (%s, %s, %s, %s, %s)"
            values = (cid, cname, caddress, email, phone)
        elif entity == "Sale":
            cursor.execute("SELECT MAX(SID) FROM Sale")
            max_sid = cursor.fetchone()[0]
            sid = max_sid + 1 if max_sid else 1
            cid = int(input("Enter Customer ID: "))
            sale_date = input("Enter Sale Date (YYYY-MM-DD): ")
            stotal_amount = float(input("Enter Total Amount: "))
            query = "INSERT INTO Sale (SID, CID, sale_date, Stotal_amount) VALUES (%s, %s, %s, %s)"
            values = (sid, cid, sale_date, stotal_amount)
        elif entity == "Sale_Details":
            cursor.execute("SELECT MAX(SDID) FROM Sale_Details")
            max_sdid = cursor.fetchone()[0]
            sdid = max_sdid + 1 if max_sdid else 1
            sid = int(input("Enter Sale ID: "))
            pid = int(input("Enter Product ID: "))
            sdquantity = int(input("Enter Quantity: "))
            query = "INSERT INTO Sale_Details (SDID, SID, PID, SDquantity) VALUES (%s, %s, %s, %s)"
            values = (sdid, sid, pid, sdquantity)
        elif entity == "Warehouse":
            cursor.execute("SELECT MAX(WID) FROM Warehouse")
            max_wid = cursor.fetchone()[0]
            wid = max_wid + 1 if max_wid else 1
            location = input("Enter Location: ")
            wcapacity = int(input("Enter Capacity: "))
            query = "INSERT INTO Warehouse (WID, location, Wcapacity) VALUES (%s, %s, %s)"
            values = (wid, location, wcapacity)
        elif entity == "Inventory_Stock":
            cursor.execute("SELECT MAX(ISID) FROM Inventory_Stock")
            max_isid = cursor.fetchone()[0]
            isid = max_isid + 1 if max_isid else 1
            pid = int(input("Enter Product ID: "))
            wid = int(input("Enter Warehouse ID: "))
            iquantity = int(input("Enter Quantity: "))
            query = "INSERT INTO Inventory_Stock (ISID, PID, WID, Iquantity) VALUES (%s, %s, %s, %s)"
            values = (isid, pid, wid, iquantity)
        else:
            print("Invalid entity.")
            return

        cursor.execute(query, values)
        db.commit()
        print(f"{entity} record created successfully.")
    except mysql.connector.Error as err:
        print(f"Error: {err}")
        db.rollback()

def read_records(entity):
    query = f"SELECT * FROM {entity}"
    try:
        cursor.execute(query)
        records = cursor.fetchall()
        for record in records:
            print(record)
    except mysql.connector.Error as err:
        print(f"Error: {err}")

def update_record(entity):
    try:
        if entity == "Product":
            pid = int(input("Enter Product ID to update: "))
            field = input("Enter field to update (Pname/Pcategory/Punit_price/Pstock_quantity): ")
            value = input("Enter new value: ")
            query = f"UPDATE Product SET {field} = %s WHERE PID = %s"
            values = (value, pid)
        elif entity == "Supplier":
            supid = int(input("Enter Supplier ID to update: "))
            field = input("Enter field to update (Sname/Saddress/Scontact_info): ")
            value = input("Enter new value: ")
            query = f"UPDATE Supplier SET {field} = %s WHERE SUPID = %s"
            values = (value, supid)
        elif entity == "Order":
            oid = int(input("Enter Order ID to update: "))
            field = input("Enter field to update (order_date/SUPID/Ototal_amount): ")
            value = input("Enter new value: ")
            query = f"UPDATE `Order` SET {field} = %s WHERE OID = %s"
            values = (value, oid)
        elif entity == "Order_Details":
            odid = int(input("Enter Order Detail ID to update: "))
            field = input("Enter field to update (OID/PID/ODquantity/ODunit_price): ")
            value = input("Enter new value: ")
            query = f"UPDATE Order_Details SET {field} = %s WHERE ODID = %s"
            values = (value, odid)
        elif entity == "Customer":
            cid = int(input("Enter Customer ID to update: "))
            field = input("Enter field to update (Cname/Caddress/Email/Phone): ")
            value = input("Enter new value: ")
            query = f"UPDATE Customer SET {field} = %s WHERE CID = %s"
            values = (value, cid)
        elif entity == "Sale":
            sid = int(input("Enter Sale ID to update: "))
            field = input("Enter field to update (CID/sale_date/Stotal_amount): ")
            value = input("Enter new value: ")
            query = f"UPDATE Sale SET {field} = %s WHERE SID = %s"
            values = (value, sid)
        elif entity == "Sale_Details":
            sdid = int(input("Enter Sale Detail ID to update: "))
            field = input("Enter field to update (SID/PID/SDquantity): ")
            value = input("Enter new value: ")
            query = f"UPDATE Sale_Details SET {field} = %s WHERE SDID = %s"
            values = (value, sdid)
        elif entity == "Warehouse":
            wid = int(input("Enter Warehouse ID to update: "))
            field = input("Enter field to update (location/Wcapacity): ")
            value = input("Enter new value: ")
            query = f"UPDATE Warehouse SET {field} = %s WHERE WID = %s"
            values = (value, wid)
        elif entity == "Inventory_Stock":
            isid = int(input("Enter Inventory Stock ID to update: "))
            field = input("Enter field to update (PID/WID/Iquantity): ")
            value = input("Enter new value: ")
            query = f"UPDATE Inventory_Stock SET {field} = %s WHERE ISID = %s"
            values = (value, isid)
        else:
            print("Invalid entity.")
            return

        cursor.execute(query, values)
        db.commit()
        print(f"{entity} record updated successfully.")
    except mysql.connector.Error as err:
        print(f"Error: {err}")
        db.rollback()

def delete_record(entity):
    try:
        if entity == "Product":
            pid = int(input("Enter Product ID to delete: "))
            # Check for references in other tables
            cursor.execute("SELECT COUNT(*) FROM Order_Details WHERE PID = %s", (pid,))
            if cursor.fetchone()[0] > 0:
                print("Cannot delete product. It is referenced in orders.")
                return
            cursor.execute("SELECT COUNT(*) FROM Sale_Details WHERE PID = %s", (pid,))
            if cursor.fetchone()[0] > 0:
                print("Cannot delete product. It is referenced in sales.")
                return
            cursor.execute("SELECT COUNT(*) FROM Inventory_Stock WHERE PID = %s", (pid,))
            if cursor.fetchone()[0] > 0:
                print("Cannot delete product. It is referenced in inventory stock.")
                return
            query = "DELETE FROM Product WHERE PID = %s"
            values = (pid,)
        elif entity == "Supplier":
            supid = int(input("Enter Supplier ID to delete: "))
            cursor.execute("SELECT COUNT(*) FROM `Order` WHERE SUPID = %s", (supid,))
            if cursor.fetchone()[0] > 0:
                print("Cannot delete supplier. It is referenced in orders.")
                return
            query = "DELETE FROM Supplier WHERE SUPID = %s"
            values = (supid,)
        elif entity == "Order":
            oid = int(input("Enter Order ID to delete: "))
            # Delete related Order_Details first
            cursor.execute("DELETE FROM Order_Details WHERE OID = %s", (oid,))
            query = "DELETE FROM `Order` WHERE OID = %s"
            values = (oid,)
        elif entity == "Order_Details":
            odid = int(input("Enter Order Detail ID to delete: "))
            query = "DELETE FROM Order_Details WHERE ODID = %s"
            values = (odid,)
        elif entity == "Customer":
            cid = int(input("Enter Customer ID to delete: "))
            cursor.execute("SELECT COUNT(*) FROM Sale WHERE CID = %s", (cid,))
            if cursor.fetchone()[0] > 0:
                print("Cannot delete customer. It is referenced in sales.")
                return
            query = "DELETE FROM Customer WHERE CID = %s"
            values = (cid,)
        elif entity == "Sale":
            sid = int(input("Enter Sale ID to delete: "))
            # Delete related Sale_Details first
            cursor.execute("DELETE FROM Sale_Details WHERE SID = %s", (sid,))
            query = "DELETE FROM Sale WHERE SID = %s"
            values = (sid,)
        elif entity == "Sale_Details":
            sdid = int(input("Enter Sale Detail ID to delete: "))
            query = "DELETE FROM Sale_Details WHERE SDID = %s"
            values = (sdid,)
        elif entity == "Warehouse":
            wid = int(input("Enter Warehouse ID to delete: "))
            cursor.execute("SELECT COUNT(*) FROM Inventory_Stock WHERE WID = %s", (wid,))
            if cursor.fetchone()[0] > 0:
                print("Cannot delete warehouse. It is referenced in inventory stock.")
                return
            query = "DELETE FROM Warehouse WHERE WID = %s"
            values = (wid,)
        elif entity == "Inventory_Stock":
            isid = int(input("Enter Inventory Stock ID to delete: "))
            query = "DELETE FROM Inventory_Stock WHERE ISID = %s"
            values = (isid,)
        else:
            print("Invalid entity.")
            return

        cursor.execute(query, values)
        db.commit()
        print(f"{entity} record deleted successfully.")
    except mysql.connector.Error as err:
        print(f"Error: {err}")
        db.rollback()

def main():
    entities = ["Product", "Supplier", "Order", "Order_Details", "Customer", "Sale", "Sale_Details", "Warehouse", "Inventory_Stock"]
    
    while True:
        print("\nInventory Management System")
        print("1. Create Record")
        print("2. Read Records")
        print("3. Update Record")
        print("4. Delete Record")
        print("5. Exit")
        choice = input("Enter your choice (1-5): ")

        if choice == '5':
            break

        print("\nChoose an entity:")
        for i, entity in enumerate(entities, 1):
            print(f"{i}. {entity}")
        entity_choice = int(input("Enter your choice (1-9): "))
        entity = entities[entity_choice - 1]

        if choice == '1':
            create_record(entity)
        elif choice == '2':
            read_records(entity)
        elif choice == '3':
            update_record(entity)
        elif choice == '4':
            delete_record(entity)
        else:
            print("Invalid choice. Please try again.")

    db.close()
    print("Program terminated.")

if __name__ == "__main__":
    main()
```
